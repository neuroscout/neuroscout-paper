# Face-related features
A naive model that codes for the presence vs. absence of faces fails to identify robust FFA activation. We hypothesis that this reflects adaptation effects: the FFA is primarily responsive to the initial presentation of an entirely new face, or to the presentation of a previously seen face after a long delay. We test this by fitting separate models that include regressors coding for (a) first-time presentation of a face, (b) cumulative time each individual face has been seen for, and (c) time elapsed since a given face last appeared on-screen.

```python
from pathlib import Path
import json
json_data = json.load((Path('../models') / 'facenet.json').open())
```

```python
any_faces_imgs = analyze_collection(json_data['any_faces'], exclude_datasets=['studyforrest', 'HealthyBrainNetwork', 'NaturalisticNeuroimagingDatabase'])
```

```python
faces_2_imgs = analyze_collection(json_data['any_faces+speech+shot_change'])
```

```python
from nilearn import plotting as niplt
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(3, 1, figsize=(12, 3 * 2.5))
plot_kwargs=dict(vmax=14, threshold=2.58, cut_coords=[-24, 2, 28, 54], display_mode='z')
niplt.plot_stat_map(any_faces_imgs[0], axes=axes[0], title='faces', **plot_kwargs)
niplt.plot_stat_map(faces_2_imgs[0], axes=axes[1], title='+ speech + shot change', **plot_kwargs)
niplt.plot_stat_map(faces_3_imgs[0], axes=axes[2], title = '+ adaptation', **plot_kwargs)
```
