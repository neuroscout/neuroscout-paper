# Meta-Analysis

The `analyze_collection` function takes a set of Neuroscout analysis IDs which reflect an equivalent statstical model, and compiles them into a `nimare` dataset for meta-analysis. 

`nimare` downloads the required inputs from every dataset level Neuroscout analysis (i.e. group level effect and variance maps) from NeuroVault, and performs a `DerSimonianLaird` image-based meta-analysis (ibma). For convenience, this function also plots resulting meta-analytic maps using `nilearn`.

```python
for label in ['brightness', 'shot_change']:
    analyze_collection(
        single_predictors[label], 
        collection_kwargs={'estimator': 'afni'},
        plot_kwargs=dict(vmax=18, cut_coords=range(-25, 55, 16), threshold=3.29)
    )
```

```python
auditory = ['speech', 'rms']
for label in auditory:
    analyze_collection(
        single_predictors[label], 
        collection_kwargs={'estimator': 'afni'},
        plot_kwargs=dict(vmax=18, cut_coords=range(-25, 55, 16), threshold=3.29)
    )
```

```python
clarifai = ['text',  'landscape', 'tool']
for label in clarifai:
    analyze_collection(
        single_predictors[label], 
        collection_kwargs={'estimator': 'afni'},
        plot_kwargs=dict(vmax=18, cut_coords=range(-25, 55, 16), threshold=3.29)
    )
```
